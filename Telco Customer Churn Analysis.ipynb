## Replacing blanks with 0 as tenure is 0 and no total charges are recorded.

```python
df["TotalCharges"]= df["TotalCharges"].replace(" ",0)
df["TotalCharges"]= df["TotalCharges"].astype("float")
```

we have converted 0 and 1 values of senior citizen to yes/no

```python
def conv(value):
    if value==1:
        return "yes"
    else: 
        return "no"
    
    

df['SeniorCitizen']= df["SeniorCitizen"].apply(conv)
df.head(30)
```

```python
ax= sns.countplot(x= 'Churn', data=df)    
ax.bar_label(ax.containers[0])
plt.title("Count of customers by Churn")
plt.show()             

```

```python
plt.figure(figsize=(3,4))
gb= df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'],labels=gb.index, autopct= "%1.2f%%" )
plt.title("Percentage of Churned Customers", fontsize=10)
plt.show()
```

From the given pie-chart we can conclude that 26.54% customers have been churned out.

Now, lets explore the reason behind it.

```python
plt.figure(figsize=(5,4))

sns.countplot(x="gender", data=df, hue="Churn")
plt.title("Churn by gender")
plt.show()
```

```python
plt.figure(figsize=(3,3))

ax= sns.countplot(x="SeniorCitizen", data=df, hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn by Senior Citizen")
plt.show()

```

Comparatively a greater percentage of people in senior citizen category have churned out.

```python
plt.figure(figsize=(6,5))
sns.histplot(x="tenure", data=df, bins=72, hue="Churn")
```

People who have used our services for a long time have stayed and people who have used our services for one or two months have churned.

```python
plt.figure(figsize=(4,4))

ax= sns.countplot(x="Contract", data=df, hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn by Contract")
plt.show()  
```

People who have month-to-month contract are likely to churn out as compare to those who have 1 or 2 year contract.

```python

# Suppose your dataframe is called df
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create subplots: let's arrange them in a grid
n_cols = 3   # number of plots per row
n_rows = (len(cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(8,8))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, ax=axes[i], palette="Set2", hue=df["Churn"])
    axes[i].set_title(f"Count of {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x', rotation=30)  # rotate x labels if needed

# Hide empty subplots (if any)
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

1. Phone Service

Majority of customers have Phone Service.

Churn is noticeably higher among customers who do have Phone Service (since most customers fall into this group).

2. Multiple Lines

Customers without phone service are few.

Both "Yes" and "No" groups show churn, but churn is somewhat higher among those with multiple lines.

3. Internet Service

DSL: Moderate churn.

Fiber Optic: Very high churn → this is the riskiest group.

No Internet: Very low churn (customers without internet are stable).

4. Online Security

Customers with No Online Security have the highest churn.

Those with security are more likely to stay, showing its protective effect.

5. Online Backup

Churn is higher among those with No Online Backup.

Having backup reduces churn likelihood.

6. Device Protection

Again, churn is much higher among customers with No protection.

Customers who subscribed are more loyal.

7. Tech Support

Strong trend: No Tech Support → High churn.

Customers with tech support show significantly lower churn.

8. Streaming TV & Streaming Movies

Subscriptions are fairly balanced between Yes/No.

Churn rates are similar across both groups.

→ These services do not significantly impact churn.
